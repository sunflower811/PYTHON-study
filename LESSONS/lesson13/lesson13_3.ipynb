```python
import numpy as np
import matplotlib.pyplot as plt

index=[3,4,5,6,7]
sub=['apple','banana','orange','tomato','guava']
values=[5000,7000,3000,4000,6000]
figure=plt.figure()
axis=figure.add_subplot(1,1,1)
axis.bar(index,values)
axis.set_xticks(index,labels=sub)
xmin,xmax,ymin,ymax=axis.axis()
xmin-=0.2
xmax+=0.2
ymin=2000
ymax+=1000
axis.axis(xmin=xmin,xmax=xmax,ymin=ymin,ymax=ymax)
for idy,x in enumerate(index):
    axis.annotate(str(values[idy]),(x-0.2,values[idy]+100))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

index=np.array([3,4,5,6,7])
sub=['apple','banana','orange','tomato','guava']
values1=[5000,7000,3000,4000,6000]
values2=[3000,4000,5000,6000,7000]
values3=[7000,6000,5000,4000,3000]
figure=plt.figure()
axis=figure.add_subplot(1,1,1)
bin_width=0.2
axis.bar(index,values1,width=bin_width,color='red')
axis.bar(index+bin_width,values2,width=bin_width,color='green')
axis.bar(index-bin_width,values3,width=bin_width)
axis.set_xticks(index,labels=sub)
xmin,xmax,ymin,ymax=axis.axis()
xmin-=0.2
xmax+=0.2
ymin=2000
ymax+=1000
axis.axis(xmin=xmin,xmax=xmax,ymin=ymin,ymax=ymax)

plt.show()
```
