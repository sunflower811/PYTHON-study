```python
import numpy as np
import matplotlib.pyplot as plt

index=['apple','banana','orange','tomato','guava']
values=[5,7,3,4,6]
figure=plt.figure()
axis=figure.add_subplot(1,1,1)
axis.bar(index,values)
print(axis.get_xticks())
print(axis.get_xticklabels())
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

index=[3,4,5,6,7]
sub=['apple','banana','orange','tomato','guava']
values=[5000,7000,3000,4000,6000]
figure=plt.figure()
axis=figure.add_subplot(1,1,1)
axis.bar(index,values)
axis.set_xticks(index,labels=sub)
xmin,xmax,ymin,ymax=axis.axis()
xmin-=0.2
xmax+=0.2
ymin=2000
ymax+=1000
axis.axis(xmin=xmin,xmax=xmax,ymin=ymin,ymax=ymax)
plt.show()
```
