```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-3,3,num=50)
y1=x*2+1
y2=x*x

fig=plt.figure()
axis=fig.add_subplot()
axis.plot(x,y1,color='red',linestyle='--')
axis.plot(x,y2,color='blue',linestyle='--')
#print(axis.get_xaxis())
axis.set_xlim(-1,2)
axis.set_ylim(-2,3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

t=np.arange(0,2.5,0.1)
y=np.sin(math.pi*t)
y2=np.sin(math.pi*t+math.pi)


figure=plt.figure()
axis=figure.add_subplot()
axis.plot(t,y,'b*')
axis.plot(t,y2,'g^')
plt.show()
```
