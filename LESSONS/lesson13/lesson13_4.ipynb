```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels=['Nokia','Samsung','Apple','Lumia']
values=[20,30,45,10]
colors=['yellow','green','red','blue']
explode=(0.2,0,0,0)
```

```python
fig=plt.figure()
axis=fig.add_subplot()
axis.pie(values,
         labels=labels,
         autopct='%1.1f%%',
         colors=colors,
         radius=0.8,
         shadow=True,
         startangle=180,
         explode=explode)

plt.show()
```
