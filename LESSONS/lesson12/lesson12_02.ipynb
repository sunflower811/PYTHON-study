```python
import matplotlib.pyplot as plt
# step1 figure實體-->畫布
# step2 axes-->圖表
fig=plt.figure()
axes=fig.add_subplot(1,1,1)
subjects=['Chinese','English','Math','History','Discovery']
nums1=[1,4,9,16,25]
nums2=[2,7,3,21,3]
axes.plot(subjects,nums1,'+--y',label='student1')
axes.plot(subjects,nums2,'+--',label='student2')
axes.set_title('students')
axes.set_xlabel('subjects')
axes.set_ylabel('scores')
axes.grid(axis='y')
axes.set_xticks([0,1,2,3,4])
axes.set_xticklabels(['ch','en','ma','hi','cover'])
axes.legend()
plt.show()
```
