```python
import matplotlib.pyplot as plt
# step1 figure實體-->畫布
# step2 axes-->圖表
fig=plt.figure()
ax_1=fig.add_subplot(1,1,1)
X=[0,1,2]
Y=[3,4,5]
ax_1.plot(X,Y,'ro-')
plt.show()
```
