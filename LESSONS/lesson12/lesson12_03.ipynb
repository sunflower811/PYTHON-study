```python
import random
days=[num for num in range(1,31)]
celsius=[round(random.uniform(28.5,35.2)) for _ in range(1,31)]
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure(figsize=(10,5))
axes=fig.add_subplot(1,1,1)

x=np.arange(1,31,1)

axes.plot(days,celsius,'o--r')
axes.set_title('july weather')
axes.set_xlabel('Days')
axes.set_ylabel('Celsius')
axes.grid()
a=[num for num in range(28,36)]
axes.set_yticks(a)
axes.set_xticks(days)
axes.legend()
plt.show()
```
